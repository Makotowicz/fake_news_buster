## Using NLTK Sentiment Analyzer

**VADER**
VADER breaks down sentiment intensity scores into a positive, negative and neutral component, which are then normalized and squashed to be within the range [-1, 1] as a “compound” score

positive sentiment : (compound score >= 0.05)
neutral sentiment : (compound score > -0.05) and (compound score < 0.05)
negative sentiment : (compound score <= -0.05)

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
vader = SentimentIntensityAnalyzer()
```

```python
first_sentence = data_tb["title"][0]
first_sentence
```

```python
vader.polarity_scores(first_sentence)
```

**-->Better version of vader sentiment functions (without 4 helper-function)**

```python
def feature_vader_polarity_scores(df, column):
    new_column_name = f'{column}_Vader_negative_score'
    df[new_column_name] = df[column].apply(lambda x: vader.polarity_scores(x)["neg"])
    
    new_column_name = f'{column}_Vader_neutral_score'
    df[new_column_name] = df[column].apply(lambda x: vader.polarity_scores(x)["neu"])
    
    new_column_name = f'{column}_Vader_positive_score'
    df[new_column_name] = df[column].apply(lambda x: vader.polarity_scores(x)["pos"])
    
    new_column_name = f'{column}_Vader_compound_score'
    df[new_column_name] = df[column].apply(lambda x: vader.polarity_scores(x)["compound"])
    
    return df
```

**Functions for Vader scores**

```python
def vader_neg_score(x):
    return vader.polarity_scores(x)["neg"]
```

```python
def vader_neu_score(x):
    return vader.polarity_scores(x)["neu"]
```

```python
def vader_pos_score(x):
    return vader.polarity_scores(x)["pos"]
```

```python
def vader_compound_score(x):
    return vader.polarity_scores(x)["compound"]
```

```python
def feature_vader_polarity_scores(df, column):
    new_column_name = f'{column}_Vader_negative_score'
    df[new_column_name] = df[column].apply(vader_neg_score)
    
    new_column_name = f'{column}_Vader_neutral_score'
    df[new_column_name] = df[column].apply(vader_neu_score)
    
    new_column_name = f'{column}_Vader_positive_score'
    df[new_column_name] = df[column].apply(vader_pos_score)
    
    new_column_name = f'{column}_Vader_compound_score'
    df[new_column_name] = df[column].apply(vader_compound_score)
    
    return df
```
